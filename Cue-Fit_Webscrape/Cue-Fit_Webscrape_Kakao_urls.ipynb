```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import re
```

```python
# 엑셀 파일 로드
exe = pd.read_excel('seoul_gym.xlsx')
```

```python
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options as ChromeOptions

chrome_options = ChromeOptions()
chrome_options.add_argument("--user-agent=Your User Agent String")

driver = webdriver.Chrome('C:/chromedriver.exe', options=chrome_options)
driver.implicitly_wait(10)
```

```python
rating_selector = "#info\.search\.place\.list > li > div.rating.clickArea > span.score > em"
review_count_selector = "#info\.search\.place\.list > li > div.rating.clickArea > span.score > a"
```

```python
# 리뷰 링크 가져오기
links = "#info\.search\.place\.list > li > div.rating.clickArea > span.score > a"
kakao_review_urls = []
ratings = []
review_counts = []
EXE200 = exe.iloc[1:202]


for center_name in EXE200['센터검색명'] :
    rating = "별점 없음"
    review_count = "후기 없음"
    try :
        driver.get('https://map.kakao.com/') # 카카오 지도 
        # 검색창에 센터 이름 입력
        search_box = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "#search\.keyword\.query"))
        )
        search_box.clear()
        search_box.send_keys(center_name)
        search_box.send_keys(Keys.RETURN)
        time.sleep(3)  # 검색 결과 로딩 대기
        # 평점 가져오기

        rating_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, rating_selector))
        )
        #rating_element = driver.find_element(By.CSS_SELECTOR, rating_selector)
        rating = rating_element.text if rating_element else "별점 없음"

        # 후기 개수 가져오기
        review_count_element = WebDriverWait(driver, 10).until(
                                EC.presence_of_element_located((By.CSS_SELECTOR, review_count_selector))
        )
        #review_count_element = driver.find_element(By.CSS_SELECTOR, review_count_selector)
        review_count = review_count_element.text if review_count_element else "후기 없음"


        elements = driver.find_elements(By.CSS_SELECTOR,links)
        if elements:
            kakao_review_url = elements[0].get_attribute("href")
        else:
            kakao_review_url = "No URL found"
        
    except Exception as error:
        print(error)
        time.sleep(2)
        kakao_review_url = "Error"
    
    kakao_review_urls.append(kakao_review_url)
    ratings.append(rating)
    review_counts.append(review_count)
        
driver.quit()

Center29 = EXE200['센터검색명']

reviews_df = pd.DataFrame({'Center Name' : Center29,'Review URL' :kakao_review_urls,'Rating':ratings, 'ReviewCounts' : review_counts})
reviews_df.to_excel('kakao_urls.xlsx', index=False)
```
