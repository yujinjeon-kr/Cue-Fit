```python
from jamo import h2j, j2hcj

# 사용자 정의 사전에 단어 추가 시, 마지막 글자의 종성 유무 규명 함수
def get_jongsung_TF(sample_text):
    sample_text_list = list(sample_text)
    last_word = sample_text_list[-1]
    last_word_jamo_list = list(j2hcj(h2j(last_word)))
    last_jamo = last_word_jamo_list[-1]

    jongsung_TF = "T"

    if last_jamo in ['ㅏ', 'ㅑ', 'ㅓ', 'ㅕ', 'ㅗ', 'ㅛ', 'ㅜ', 'ㅠ', 'ㅡ', 'ㅣ', 'ㅘ', 'ㅚ', 'ㅙ', 'ㅝ', 'ㅞ', 'ㅢ', 'ㅐ,ㅔ', 'ㅟ', 'ㅖ', 'ㅒ']:
        jongsung_TF = "F"

    return jongsung_TF
```

```python
# 사용자단어 사전 'user-nnp.csv' 파일 생성 함수
# 단어 목록에 포함된 단어의 끝글자 종성유무 판단 후 file_data 리스트에 추가
# file_data 리스트의 각 줄을 읽어 user-nnp.csv 파일에 입력

def make_user_dic_csv(morpheme_type, word_list, user_dic_file_name):
  file_data = []

  for word, score in word_list:
    jongsung_TF = get_jongsung_TF(word)

    line = f"{word},,,{score},{morpheme_type},*,{jongsung_TF},{word},*,*,*,*,*\n"

    file_data.append(line)

  with open("./user-dic/user-nnp.csv", 'w', encoding='utf-8') as f:
    for line in file_data:
      f.write(line)
```

```python
import numpy as np
import pandas as pd
import warnings # 경고 메시지 무시
warnings.filterwarnings(action='ignore')
from konlpy.tag import Mecab
mecab = Mecab()
from collections import Counter # 단어 등장 횟수 카운트
```

```python
# Mecab 초기화
mecab = MeCab.Tagger()

# 명사 추출 함수 정의
def extract_nouns(text):
    node = mecab.parseToNode(text)
    nouns = []
    while node:
        if node.feature.split(',')[0] == 'NNG' or node.feature.split(',')[0] == 'NNP':
            nouns.append(node.surface)
        node = node.next
    return nouns

# '본문' 열의 각 행에 대해 명사 추출하여 '명사 토큰' 열 생성
jambaekee['명사 토큰'] = jambaekee['본문'].apply(extract_nouns)

# 결과 출력
print(jambaekee[['본문', '명사 토큰']])
```
