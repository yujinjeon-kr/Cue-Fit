```python
# 카테고리 사전 로드
category = pd.read_excel('/content/gdrive/MyDrive/빅데이터/파이널/NLP/최종LDA/1215_최종_카테고리사전.xlsx')
category.head()
```

```python
# 딕셔너리 생성
category_keyword_dict = {}

for _, row in category.iterrows():
    category = row['Category']
    keyword = row['Keyword']

    if category not in category_keyword_dict:
        category_keyword_dict[category] = [keyword]
    else:
        category_keyword_dict[category].append(keyword)

# 결과 출력
print(category_keyword_dict)
```

```python
def check_duplicate_values(dictionary):
    values_set = set()

    for values_list in dictionary.values():
        for value in values_list:
            if value in values_set:
                return False
            values_set.add(value)

    return True

# 중복 확인
result = check_duplicate_values(category_keyword_dict)

# 결과 출력
if result:
    print("딕셔너리의 값 중에 중복이 없습니다.")
else:
    print("딕셔너리의 값 중에 중복이 있습니다.")
```

```python
# 각 카테고리별 중복 제거
for category in category_keyword_dict:
    category_keyword_dict[category] = list(set(category_keyword_dict[category]))

# 전체 카테고리 키워드 통합
all_keywords = []
for keywords in category_keyword_dict.values():
    all_keywords.extend(keywords)

# 전체 중복 키워드 찾기
duplicates = set([keyword for keyword in all_keywords if all_keywords.count(keyword) > 1])
duplicates
```

```python
# 각 카테고리에서 중복 키워드 제거
for category in category_keyword_dict:
    category_keyword_dict[category] = [keyword for keyword in category_keyword_dict[category] if keyword not in duplicates]

print(category_keyword_dict)
```

```python
# 사용자 정의 사전에 단어를 추가할 때, 해당 단어의 종성 유무에 따라 발음이나 문법적 특성을 정의하는 데 사용
from jamo import h2j, j2hcj


def get_jongsung_TF(sample_text):
    sample_text_list = list(sample_text)
    last_word = sample_text_list[-1]
    last_word_jamo_list = list(j2hcj(h2j(last_word)))
    last_jamo = last_word_jamo_list[-1]

    jongsung_TF = "T"

    if last_jamo in ['ㅏ', 'ㅑ', 'ㅓ', 'ㅕ', 'ㅗ', 'ㅛ', 'ㅜ', 'ㅠ', 'ㅡ', 'ㅣ', 'ㅘ', 'ㅚ', 'ㅙ', 'ㅝ', 'ㅞ', 'ㅢ', 'ㅐ,ㅔ', 'ㅟ', 'ㅖ', 'ㅒ']:
        jongsung_TF = "F"

    return jongsung_TF
```

```python
def make_user_dic_csv(morpheme_type, word_list, user_dic_file_name):
  file_data = []

  for word, score in word_list:
    jongsung_TF = get_jongsung_TF(word)

    line = f"{word},,,{score},{morpheme_type},*,{jongsung_TF},{word},*,*,*,*,*\n"

    file_data.append(line)

  with open("./user-dic/user-nnp.csv", 'w', encoding='utf-8') as f:
    for line in file_data:
      f.write(line)
```

```python
# Mecab 형태소 분석기를 사용하여 명사, 어근, 형용사, 동사를 추출
def extract_tokens(text):
    pos = mecab.pos(text)
    tokens = [word for word, tag in pos if tag in ['NNG', 'NNP', 'VV', 'VA', 'XR']]
    return tokens

data['tokenized'] = list(map(extract_tokens, data['Review Text']))
data['tokenized']
```

# 2. 치환

```python
# 단어를 치환하는 함수 정의
def replace_words(word_list):
    return [replace_dict[word] if word in replace_dict else word for word in word_list]
```

```python
# df의 'replacement' 열에 치환 함수 적용
data['replacement'] = data['tokenized'].apply(replace_words)

# 결과 확인
print(data['replacement'])
```

# 5. 바이그램 생성

```python
# 리스트 형태의 단어들로부터 2-gram 생성 함수
def create_ngrams_or_return_list(word_list, n=2):
    # 리스트 길이가 n보다 작으면 원래 리스트 반환
    if len(word_list) < n:
        return word_list
    # 2-gram 생성
    return [' '.join(grams) for grams in ngrams(word_list, n)]

# 'stopword_removed' 열에 2-gram 적용
data['바이그램'] = data['불용어제거'].apply(lambda x: create_ngrams_or_return_list(x, 2))

# 결과 확인
print(data[['바이그램']])
```

```python
import matplotlib.pyplot as plt
from gensim.models.ldamodel import LdaModel
from gensim.models.coherencemodel import CoherenceModel

# Coherence Score 계산을 위한 함수
def compute_coherence_values(dictionary, corpus, texts, start, limit, step):
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        model = LdaModel(corpus=corpus, num_topics=num_topics, id2word=dictionary, passes=10)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary,
                                        coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())
    return model_list, coherence_values

# LDA 모델과 Coherence Score 계산
start = 2
limit = 15
step = 1
model_list, coherence_values = compute_coherence_values(dictionary=dictionary_gensim,
                                                        corpus=corpus_gensim, texts=texts,
                                                        start=start, limit=limit, step=step)

# 토픽 수와 Coherence Score를 기반으로 그래프 그리기
x = range(start, limit, step)
plt.plot(x, coherence_values)
plt.xlabel("토픽개수")
plt.ylabel("Coherence score")
plt.title("토픽개수에 따른 Coherence Score")
plt.show()
```
